```python
lc = GCRCatalogs.load_catalog('dc2_truth_run1.2_variable_lightcurve')
lc.list_all_quantities(include_native=True)
```

# Testing

```python
n_plotted = 0
for q in lc.get_quantities(['uniqueId','mjd', 'mag','filter'], native_filters=['sn == 1', 'ra > 40'], return_iterator=True):
    if len(q['mjd']) >= 20:
        plt.plot(q['mjd'][q['filter']==3], q['mag'][q['filter']==3], '.-')
        n_plotted += 1
    if n_plotted >= 3:
        break
# plt.ylim(35, 24)
plt.xlabel('MJD')
plt.ylabel('$i$ [mag]')
```

# Curve Fitting

```python
import os
```

```python
wdir = '/global/u1/d/djp81/public'
file_list = os.listdir('/global/u1/d/djp81/public')
# print(file_list)
```

```python
s = snpy.get_sn(wdir+'/'+file_list[0])

s.choose_model('max_model')
s.fit()

s.plot()

s.summary()
```

```python
d_mod = []
z = []

for i in range(100):
    if i%50==0: print i
    file = file_list[i]
    try:
        s = snpy.get_sn(wdir+'/'+file)
        s.choose_model('max_model')
    
        s.fit()
        
    except (RuntimeError,ValueError,IndexError,KeyError) as e :
        print('error',e,file)
        continue
    
    d_mod.append(s.get_distmod(cosmo='LambdaCDM'))
    
    with open(wdir+'/'+file) as f:
        z.append(f.readline().split(' ')[1])
        
plt.semilogy(z,d_mod,'.')
plt.xlabel('redshift')
plt.ylabel('d_mod')
```

```python
z = [float(i) for i in z]
```
